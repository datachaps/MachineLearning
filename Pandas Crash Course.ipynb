## Statistics

```python
# descriptive statistics, applicabe only for numeric fields
df.describe()
```

```python
# statistics for individual columns
print(df['Mileage'].mean()) # average
print(df['Mileage'].std()) # standard deviation
print(df['Mileage'].min())
print(df['Mileage'].quantile(0.25)) # also called Q1
print(df['Mileage'].median())       # also called Q2
print(df['Mileage'].quantile(0.75)) # also called Q3
print(df['Mileage'].max())
```

## Filtering

```python
df_filtered = df[df['Brand'] == 'Audi']
df_filtered
```

```python
df_filtered = df[(df['Brand'] == 'Audi') | (df['Brand'] == 'Mercedes')]
df_filtered
```

```python
df_filtered = df[df['Brand'].str.contains('Merc')]
df_filtered
```

```python
df_filtered = df[df['Mileage'] > 290000]
df_filtered
```

```python
df_filtered = df.query("Brand == 'Audi' & Doors > 4 & Mileage > 70000")
df_filtered
```
